## Who is our best Customer? (RFM analysis)

```python
%%sql

select CUSTOMERNAME,
    sum(sales) MonetaryValue,
	avg(sales) AvgMonetaryValue,
	count(ORDERNUMBER) Frequency,
	TIMESTAMPDIFF(DAY, MAX(ORDER_DATE), (SELECT max(ORDER_DATE) FROM sales_data_sample)) Recency
from sales_data_sample
group by CUSTOMERNAME
```

## Grouping the customers

```python
%%sql

-- Drop table if it exists
DROP TABLE IF EXISTS rfm_temp;

-- Create RFM calculations and store in a temporary table
CREATE TEMPORARY TABLE rfm_temp AS
WITH rfm AS (
    SELECT CUSTOMERNAME,
        SUM(sales) AS MonetaryValue,
        AVG(sales) AS AvgMonetaryValue,
        COUNT(ORDERNUMBER) AS Frequency,
        DATEDIFF(CURDATE(), MAX(ORDER_DATE)) AS Recency
    FROM sales_data_sample
    GROUP BY CUSTOMERNAME
)
SELECT r.*,
    NTILE(4) OVER (ORDER BY Recency) AS rfm_recency,
    NTILE(4) OVER (ORDER BY Frequency) AS rfm_frequency,
    NTILE(4) OVER (ORDER BY MonetaryValue) AS rfm_monetary
FROM rfm r;
```

```python
%%sql

-- View the data in the rfm_temp table
SELECT * FROM rfm_temp;
```

```python
%%sql

-- Add RFM Cell and RFM Cell String to the temporary table
ALTER TABLE rfm_temp
ADD COLUMN rfm_cell INT,
ADD COLUMN rfm_cell_string VARCHAR(10);

UPDATE rfm_temp
SET rfm_cell = (rfm_recency + rfm_frequency + rfm_monetary),
    rfm_cell_string = CONCAT(CAST(rfm_recency AS CHAR), CAST(rfm_frequency AS CHAR), CAST(rfm_monetary AS CHAR));
```

```python
%%sql

-- View the data in the rfm_temp table
SELECT * FROM rfm_temp;
```

```python
%%sql

-- RFM Segmentation
SELECT CUSTOMERNAME, rfm_recency, rfm_frequency, rfm_monetary, rfm_cell_string,
    CASE
        -- Lost Customers (low frequency and monetary, high recency)
        WHEN rfm_cell_string IN ('111', '112', '121', '123', '132', '211', '114', '141') THEN 'Lost Customer'
        
        -- Big spenders who haven't purchased lately (high monetary, low recency)
        WHEN rfm_cell_string IN ('133', '134', '143', '144', '244', '334', '343', '344') THEN 'Slipping Away (High Value)'

        -- New customers (low frequency but recent)
        WHEN rfm_cell_string IN ('311', '411', '331') THEN 'New Customer'

        -- Potential Churners (medium frequency, medium recency)
        WHEN rfm_cell_string IN ('222', '223', '233', '322', '221', '231') THEN 'Potential Churners'

        -- Active Customers (medium to high frequency, moderate recency)
        WHEN rfm_cell_string IN ('323', '333', '321', '422', '332', '432', '231', '312', '313') THEN 'Active'

        -- Loyal Customers (high frequency, low recency, high monetary)
        WHEN rfm_cell_string IN ('433', '434', '443', '444', '413', '414', '423', '424') THEN 'Loyal Customer'

        -- High Potential (Frequent but spending less)
        WHEN rfm_cell_string IN ('211', '212', '213', '311', '411', '221', '222', '232') THEN 'High Potential'

        -- At Risk (High spenders but not recent)
        WHEN rfm_cell_string IN ('331', '341', '431', '441', '342', '432') THEN 'At Risk'

        -- Champions (Best customers, high in all RFM)
        WHEN rfm_cell_string IN ('111', '112', '113', '122', '132') THEN 'Champion'

        -- Default group for any uncategorized values
        ELSE 'Others'
    END AS rfm_segment
FROM rfm_temp;
```

## Put the result in the database 

```python
%%sql

CREATE TABLE rfm_customer_seg AS 
SELECT * FROM rfm_temp;
```

```python
%%sql

CREATE VIEW rfm_segmentation_view AS
SELECT CUSTOMERNAME, rfm_recency, rfm_frequency, rfm_monetary, rfm_cell_string,
    CASE
        WHEN rfm_cell_string IN ('111', '112', '121', '123', '132', '211', '114', '141') THEN 'Lost Customer'
        WHEN rfm_cell_string IN ('133', '134', '143', '144', '244', '334', '343', '344') THEN 'Slipping Away (High Value)'
        WHEN rfm_cell_string IN ('311', '411', '331') THEN 'New Customer'
        WHEN rfm_cell_string IN ('222', '223', '233', '322', '221', '231') THEN 'Potential Churners'
        WHEN rfm_cell_string IN ('323', '333', '321', '422', '332', '432', '231', '312', '313') THEN 'Active'
        WHEN rfm_cell_string IN ('433', '434', '443', '444', '413', '414', '423', '424') THEN 'Loyal Customer'
        WHEN rfm_cell_string IN ('211', '212', '213', '311', '411', '221', '222', '232') THEN 'High Potential'
        WHEN rfm_cell_string IN ('331', '341', '431', '441', '342', '432') THEN 'At Risk'
        WHEN rfm_cell_string IN ('111', '112', '113', '122', '132') THEN 'Champion'
        ELSE 'Others'
    END AS rfm_segment
FROM rfm_customer_seg;

```
